**Exercise 1 (10 pts)**. Avoding catastrophic cancellation.

The tail of the standard logistic distributon is given by $1 - F(t) = 1 - (1+e^{-1})^{-1}$.

- Define a function `f1` to calculate the tail probability of the logistic distribution using the formula given above
- Use `sympy` to find the exact value of the tail distribution (using the same symbolic formula) to 20 decimal digits
- Calculate the *relative error* of `f1` when $t = 25$ (The relative error is given by `abs(exact - approximate)/exact`)
- Rewrite the expression for the tail of the logistic distribution using simple algebra so that there is no risk of cancellation, and write a function `f2` using this formula. Calculate the *relative error* of `f2` when $t = 25$. 
- How much more accurate is `f2` compared with `f1` in terms of the relative error?

````python
def f1(t):
    '''Calculate tail probability of the logistic distribution'''
    f_t = 1 - (1 + np.exp(-t))**(-1)
    return(f_t)

print('The tail probability for t = 25 is', f1(25)) # tail probability for t = 25 

from sympy import N, E
sympy_E = N(1 - (1 + E**(-25))**(-1), 20) # calculate exact tail probability for t = 25
print('The exact value of the tail probability using sympy is',sympy_E)

print('The relative error of f1 when t=25 is', abs(sympy_E - f1(25))/sympy_E)

def f2(t):
    '''Another expression to calculate tail probability of the logistric distribution'''
    f = 1/(1+np.exp(t))
    return(f)

print('The tail probability when using the new expression is', f2(25))
print('The relative error of f2 is', abs(sympy_E - f2(25))/sympy_E)
print('The relative error of f2 is much smaller compared to the relative error of f1.') 
print('This is because we have rewritten the expression to avoid cancellation.')
````

**Exercise 2 (10 pts)**. Ill-conditioned linear problems.

The solution $\beta$ can also be loaded as

- Write a formula that could solve the system of linear equations in terms of $X$ and $y$ Write a function `f1` that takes arguments $X$ and $y$ and returns $\beta$ using this formula.
- How could you code this formula using `np.linalg.solve` that does not require inverting a matrix? Write a function `f2` that takes arguments $X$ and $y$ and returns $\beta$ using this.
- Note that carefully designed algorithms *can* solve this ill-conditioned problem, which is why you should always use library functions for linear algebra rather than write your own.
```python
np.linalg.lstsq(x, y)[0]
```
- What happens if you try to solve for $\beta$ using `f1` or `f2`? Remove the column of $X$ that is making the matrix singluar and find the $p-1$ vector $b$ using `f2`.
- Note that the solution differs from that given by `np.linalg.lstsq`? This arises because the relevant condition number for `f2` is actually for the matrix $X^TX$ while the condition number of `lstsq` is for the matrix $X$. Why is the condition so high even after removing the column that makes the matrix singular?

````python
X = np.load('x.npy')
y = np.load('y.npy')
beta = np.load('b.npy')

def f1(a, b):
    '''Solve system of linear equations by (a.T*a)^(-1)*a.T*b'''
    return(np.dot(np.dot(np.linalg.inv(np.dot(a.T, a)),a.T),b))

def f2(a, b):
    '''Solve system of linear equations for non-square matrix'''
    ''' a.T*a*beta = a.T*b'''
    return(np.linalg.solve(np.dot(a.T,a), np.dot(a.T,b)))


# Remove column 5 because it causes singularity in matrix X
print(f1(X[:,0:5],y))
print(f2(X[:,0:5],y))
print(np.linalg.lstsq(X[:,0:5],y)[0])

# The condition is still high even after removing the column that makes the matrix singular because
# there is still multicolinearity among the columns. 
print(np.linalg.cond(np.dot(X[:,0:5].T,X[:,0:5])))
print(np.linalg.cond(X[:,0:5]))
````

**Exercise 3 (10 pts)**. Importance of using efficient algorihtms.

- Implement bubble sort
- Calculate its big $\mathcal{O}$ algorithmic complexity
- Time the performance of bubble sort on random uniform deviate vectors of sizes `range(100, 2000, 100)` using `time.time()` from the standard library
- Use `scipy.optimize.curve_fit` to fit an appropriate function to the collection of (size, execution time) data points. Extrapolate how long it would take to sort a random vector of size 1,000,000. Now time how long it takes for the system sort to sort a random vector of size 1,000,000.
- Plot the fits together with the data points uisng `matplotlib.pyplot` functions.

````python
def bubble_sort(A):
    """Returns a list of sorted values by comparing each pair of adjacent values and reordering them if they are in the wrong order"""
    n = len(A)
    while True:
        swapped = False
        for i in range(1, n):
            if(A[i-1] > A[i]):
                A[i-1], A[i] = A[i], A[i-1]
                swapped = True
            else:
                next
        n = n - 1 
        if(n == 0 or swapped == False):
            break
    return(A)

from numpy.random import randint
import time
np.random.seed(123)

n = range(100,2000,100)
time_bubble = []
for i in range(len(n)):
    a = randint(0, n[i], n[i])
    start_time = time.time()
    new_a = bubble_sort(a)
    time_bubble.append(time.time() - start_time)
    
print("The time (in seconds) to sort a list of random integers of various lengths:")
print(time_bubble)
````

````python
# Plot bubble sort times 
from scipy.optimize import curve_fit
def fit_sq(xdata, a, b, c):
    return a*xdata**2 + b*xdata + c

fitParams, fitCov = curve_fit(fit_sq, n, time_bubble)
plt.ylabel('Time(s)')
plt.xlabel('Size of random vector')
xfit = np.linspace(0,2000,100)
yfit = fit_sq(xfit, fitParams[0], fitParams[1], fitParams[2])
plt.plot(xfit, yfit)
plt.plot(n, time_bubble, 'bo')
````

````python
print('The time it takes to sort a list of length 1,000,000 would be %.3f seconds.' % fit_sq(1000000, fitParams[0], fitParams[1], fitParams[2]))
````

**Exercise 4 (20 pts)**. One of the goals of the course it that you will be able to implement novel algorihtms from the literature. 

- Implement the mean-shift algorithm in 1D as described [here](http://homepages.inf.ed.ac.uk/rbf/CVonline/LOCAL_COPIES/TUZEL1/MeanShift.pdf). 
    - Use the following function signature
    ```python
    def mean_shift(xs, x, kernel, max_iters=100, tol=1e-6):
    ```
    - xs is the data set, x is the starting location, and kernel is a kernel function
    - tol is the difference in $||x||$ across iterations
- Use the following kernels with bandwidth $h$ (a default value of 1.0 will work fine)
    - Flat - return 1 if $||x|| < h$ and 0 otherwise
    - Gaussian 
    $$\frac{1}{\sqrt{2 \pi h}}e^{\frac{-||x||^2}{h^2}}$$
    - Note that $||x||$ is the norm of the data point being evaluated minus the current value of $x$
- Use both kernels to find all 3 modes of the data set in `x1d.npy`
- Modify the algorithm and/or kernels so that it now works in an arbitrary number of dimensions.
- Use both kernels to find all 3 modes of the data set in `x2d.npy`
- Plot the path of successive intermediate solutions of the mean-shift algorithm starting from `x0 = (-4, 10)` until it converges onto a mode in the 2D data for each kernel. Superimpose the path on top of a contour plot of the data density.

````python
# Write a mean-shifts function for 1D data

def gauss_k(x, h = 1):
    '''Gaussian kernel'''
    return((2*np.pi*h)**(-0.5)*np.exp(-x**2 / h**2 ))
    
def flat_k(x, h = 2):
    '''Flat kernel: returns 1 if norm(x) < h and 0 otherwise'''
    '''Note that the bandwidth was tuned to 2 (instead of default 1)'''
    return([1 if abs(x) < h else 0 for x in x])

def mean_shifts(kernel, data, x, max_iters=100, tol=1e-6):
    '''Returns the point x which converges within the tolerance level'''
    diff = 1
    count = 0
    m = x
    while diff > tol:
        count += 1
        f_data = kernel(m - data)
        m_cur = sum(data*f_data)/sum(f_data)
        diff = abs(m - m_cur)
        m = m_cur
        if(count == max_iters):
            break
    return(m)
````

````python
# Apply the functions above to 'x1d.npy' for find 3 modes using the flat and gaussian kernels.
xs = np.load('x1d.npy')

RES = map(lambda k: mean_shifts(kernel = gauss_k, data = xs, x = k), xs) # iterate through n points to compute mean shifts
mode = set([round(em, 0) for em in RES]) # round the results to the neares integer and extract unique values
print(mode) # print 3 modes of the dataset using gaussian kernel
````

````python
RES_flat = map(lambda k: mean_shifts(kernel = flat_k, data = xs, x = k), xs) # iterate through n points to compute mean shifts
mode_flat = set([round(em, 0) for em in RES_flat]) # round the results to the neares integer and extract unique values
print(mode_flat) # print 3 modes of the dataset using flat kernel
````

````python
# Update flat_k and mean_shifts to use for N-dimensional data set. 
def flat_k(x, h = 1.5):
    '''Non-vectorize version of the previous flat_k version'''
    '''Note that the bandwidth was tuned to 1.5'''
    return(1 if abs(x) < h else 0)

def mean_shifts(kernel, data, x, max_iters=100, tol=1e-6):
    '''Calculate mean shifts for a vector of length n'''
    diff = [1, 1]
    count = 0
    m = x
    while np.any(diff > tol):
        count += 1
        f_data = kernel(np.linalg.norm(m - data, axis = 1))
        m_cur = (data*f_data[:, np.newaxis]).sum(axis = 0) / sum(f_data)
        diff = abs(m - m_cur)
        m = m_cur
        if(count == max_iters):
            break
    return(m)

# The following function was posted on stackoverflow by Joe Kington
# http://stackoverflow.com/questions/16970982/find-unique-rows-in-numpy-array
def unique_rows(data):
    '''Extract unique rows of a 2D array'''
    uniq = np.unique(data.view(data.dtype.descr * data.shape[1]))
    return uniq.view(data.dtype).reshape(-1, data.shape[1])
````
