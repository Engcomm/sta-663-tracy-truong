### Estimating parameters of a logistic model

Gelman's book has an example where the dose of a drug may be affected to the number of rat deaths in an experiment.

| Dose (log g/ml) | # Rats | # Deaths |
|-----------------|--------|----------|
| -0.896          | 5      | 0        |
| -0.296          | 5      | 1        |
| -0.053          | 5      | 3        |
| 0.727           | 5      | 5        |

We will model the number of deaths as a random sample from a binomial distribution, where $n$ is the number of rats and $p$ the probabbility of a rat dying. We are given $n = 5$, but we believve that $p$ may be related to the drug dose $x$. As $x$ increases the number of rats dying seems to increase, and since $p$ is a probability, we use the following model:

$$
y \sim \text{Bin}(n, p) \\
\text{logit}(p) = \alpha + \beta x \\
\alpha \sim \mathcal{N}(0, 5) \\
\beta \sim \mathcal{N}(0, 10)
$$

where we set vague priors for $\alpha$ and $\beta$, the parameters for the logistic model.

```python
# The following is written as C code to run MCMC for a logistic model in pystan
log_reg_code = """
data {
    int<lower=0> N;
    int n[N];
    int deaths[N];
    real dose[N];
}
transformed data {}
parameters {
    real alpha;
    real beta;
}
transformed parameters {}
model {
    alpha ~ normal(0, 5);
    beta ~ normal(0, 10);
    for(i in 1:N) {
        deaths[i] ~ binomial(n[i],inv_logit(alpha + beta*dose[i]));
  }
}
generated quantities {}
"""

n = np.array([5,5,5,5]) # number of mice in each dosage level
x = np.array([-0.896, -0.296, -0.053, 0.727]) # dosages for 4 groups of mice
y = np.array([0, 1, 3, 5]) # number of deaths in each group

# specify paramters for log_reg_code
log_reg_dat = {
             'N': len(n),
             'n': n,
             'deaths': y,
             'dose': x
            }

fit = pystan.stan(model_code=log_reg_code, data=log_reg_dat, iter=2000, chains=1) # run MCMC for the model provided in log_reg_code
```

```python
print 'The parameter estimates for alpha and beta from pystan are given below.'
print(fit)
```

### Using a hierarchcical model

This uses the Gelman radon data set and is based off this [IPython notebook](http://twiecki.github.io/blog/2014/03/17/bayesian-glms-3/). Radon levels were measured in houses from all counties in several states. Here we want to know if the preence of a basement affects the level of radon, and if this is affected by which county the house is located in. 

The data set provided is just for the state of Minnesota, which has 85 counties with 2 to 116 measurements per county. We only need 3 columns for this example `county`, `log_radon`, `floor`, where `floor=0` indicates that there is a basement.

We will perfrom simple linear regression on log_radon as a function of county and floor.


```python
radon = pd.read_csv('radon.csv')[['county', 'floor', 'log_radon']]
# recode county with numerical levels
radon['county_factor'] = pd.Categorical.from_array(radon.county).labels
radon['county_factor'] += 1 
# recode floor as levels to avoid floats error
radon['floor_factor'] = pd.Categorical.from_array(radon.floor).labels
radon.head()
```

```python
npr.seed(1234)
# The following is written as C code to run a hierarchical simple linear regression model in pystan
radon_hier_code = """
data {
    int<lower=0> N; // number of observations
    int<lower=0> J; // number of counties
    int<lower=1,upper=J> county[N]; 
    int<lower=0,upper=1> floor_factor[N];
    real radon[N];
}
parameters {
    vector[1] alpha[J];
    vector[1] beta[J];
    real mu_a;
    real mu_b;
    real sigma_a;
    real sigma_b;
    real sigma_y;
}
model {
    sigma_y ~ gamma(10,1);
    mu_a ~ normal(0,100);
    sigma_a ~ uniform(0,100);
    mu_b ~ normal(0,100);
    sigma_b ~ uniform(0,100);
    for(j in 1:J) {
        alpha[j] ~ normal(mu_a,sigma_a);
        beta[j] ~ normal(mu_b,sigma_b);
    }

    for(i in 1:N) {
        radon[i] ~ normal(alpha[county[i]] + beta[county[i]]*floor_factor[i],sigma_y);
    }
}

generated quantities {}
"""

# specify parameters for radon_hier_code
radon_hier_dat = {
                  'N': radon.shape[0],
                  'J': len(radon.county_factor.unique()),
                  'county': radon.county_factor,
                  'floor_factor': radon.floor_factor,
                  'radon': radon.log_radon
                }

fit = pystan.stan(model_code=radon_hier_code, data=radon_hier_dat, iter=1000, chains=1) # fit hierarchical model through pystan
```

```python
print 'There are 85 alphas and 85 betas representing the intercept and slope for each county.'
# print(fit) # uncomment to print all parameters
```

```python
ax = fit.traceplot(['alpha','beta']) 
```

**Exercise 3 (10 points)**: Implement the leap-frog method and simulate the spring dynamics following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the spring trajectory plots.

```python
npr.seed(12345)
# Leap-frog method
delta = 0.1 # step size
L = 70 # number of leap frog
def K(p):
    '''returns kinetic energy at a momentum p.'''
    return (p**2)/2

def U(x):
    '''returns potential energy of a spring at position x.'''
    return 0.5*(x**2)

def dU(x):
    '''returns the gradient of potential energy.'''
    return x

x0 = -4 # initial position
p0 = 1 # inital momentum

pStep = p0 - delta*0.5*dU(x0) # first half-step for momentum
xStep = x0 + delta*pStep # first full-step for position 

p = [] # NULL list to store leap-frog values of momentum
x = [] # NULL list to store leap-frog values of position
p.append(p0)
x.append(x0)
for i in range(0,(L-1)): # step through L-1 numbers of leap-frog
    p.append(pStep) # store momentum 
    x.append(xStep) # store position
    pStep = pStep - delta*dU(xStep) # update momentum
    xStep = xStep + delta*pStep # update position
pStep = pStep - delta*0.5*dU(xStep) # last half-step for momentum to complete an iteration of the leap-frog method
```

```python
# choose an instance within the numbers of leap-frog performed above
ind = 7
xPlot = x[ind]
pPlot = p[ind]

print 'An instance of the spring trajectory plot:'
plt.subplot(2,1,1)
xx = np.linspace(-6, xPlot, 1000)
plt.plot(xx,np.sin(6*np.linspace(0,2*np.pi,1000)),'k-')
plt.plot(xPlot,0,'bo',markersize=10)
plt.xlim([-6,6])
plt.title('Harmonic Oscillator')

ax = plt.subplot(2,2,3)
p1 = ax.bar(0, U(xPlot), 0.5, color='blue')
p2 = ax.bar(0, K(pPlot), 0.5, color='lime',bottom=U(xPlot))
p3 = ax.bar(1, U(xPlot)+K(pPlot), 0.5, color='lime')
ax.set_xticks((0.25,1.25))
ax.set_xticklabels(("U+K","H"))
plt.ylim((0,10))
plt.title("Energy")

plt.subplot(2,2,4)
plt.plot(xPlot,pPlot,'ko',markersize=10)
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.title('Phase Space')
plt.xlabel('x')
plt.ylabel('p')
```

```python
# choose an instance within the numbers of leap-frog performed above
ind = 18
xPlot = x[ind]
pPlot = p[ind]

print 'Another instance of the spring trajectory plot: note the difference in position, energy, and momentum from the previous plot.'
plt.subplot(2,1,1)
xx = np.linspace(-6, xPlot, 1000)
plt.plot(xx,np.sin(6*np.linspace(0,2*np.pi,1000)),'k-')
plt.plot(xPlot,0,'bo',markersize=10)
plt.xlim([-6,6])
plt.title('Harmonic Oscillator')

ax = plt.subplot(2,2,3)
p1 = ax.bar(0, U(xPlot), 0.5, color='blue')
p2 = ax.bar(0, K(pPlot), 0.5, color='lime',bottom=U(xPlot))
p3 = ax.bar(1, U(xPlot)+K(pPlot), 0.5, color='lime')
ax.set_xticks((0.25,1.25))
ax.set_xticklabels(("U+K","H"))
plt.ylim((0,10))
plt.title("Energy")

plt.subplot(2,2,4)
plt.plot(xPlot,pPlot,'ko',markersize=10)
plt.xlim([-6,6])
plt.ylim([-6,6])
plt.title('Phase Space')
plt.xlabel('x')
plt.ylabel('p')
```

**Exercise 4 (20 points)**: Now implement the HMC following the example at https://theclevermachine.wordpress.com/2012/11/18/mcmc-hamiltonian-monte-carlo-a-k-a-hybrid-monte-carlo/ and replicate the bivariate normal MCMC plots. (For another simple explanation of HMC, see Chapter 14 of "Doing Basyesian Data Analysis" by Jouh Kruschke.) 

```python
import scipy.linalg as la

# Implement HMC for bivariate normal
# the target distribution is p(x) = N(mu, sigma), where mu = [mu1, mu2] = [0,0] and sigma = np.array([[1.,0.8],[0.8,1.]])
npr.seed(123456) # set seed
delta = 0.3 # step size
n = 1000 # numbers of samples
L = 20 # numbers of leap-frog

def U(x, sigma):
    '''returns potential energy.'''
    return(np.dot(x.T,np.dot(la.inv(sigma),x)))

def dU(x, sigma):
    '''returns gradient of potential energy.'''
    return(np.dot(x.T,la.inv(sigma)))

def K(p):
    '''returns kinetic energy.'''
    return(0.5*np.dot(p.T,p))

sigma = np.array([[1.,0.8],[0.8,1.]]) # covariance matrix of the bivariate normal target distribution
x = np.empty((2,n+1)) # an empty array to store samples from p(x)
x0 = np.array([[0.],[6.]]) # initial values of x
x[:,0] = x0.T # store x0

t = 0 # sample counter
while t < n:
    t += 1
    p0 = npr.normal(0,1,size=(1,2)) # sample random momentum from a standard normal distribution
    pStar = p0 - delta*0.5*dU(x[:,t-1],sigma) # first half-step for momentum
    xStar = x[:,t-1] + delta*pStar # first full-step for position

    for i in range(0,(L-1)): # iterate through L-1 numbers of leap-frog
        pStar -= delta*dU(xStar.T,sigma) # update momentum
        xStar += delta*pStar # update position
    
    pStar -= delta*0.5*dU(xStar.T,sigma) # last half-step for momentum
    
    U0 = U(x[:,t-1],sigma) # evaluate potential energy at the start of trajectory
    UStar = U(xStar.T,sigma) # evaluate potential energy at the end of trajectory

    K0 = K(p0.T) # evaluate kinetic energy at the start of trajectory
    KStar = K(pStar.T) # evaluate kinetic energy at the end of trajectory

    alpha = np.min([1.,np.exp((U0+K0)-(UStar+KStar))]) # acceptance criterion
    u = npr.uniform(0,1)
    if(u < alpha):
        x[:,t] = xStar 
    else:
        x[:,t] = x[:,t-1]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x[0,:],x[1,:],label='samples')
plt.plot(x[0,0:50],x[1,0:50],'o-',label='1st 50 states',markersize=8,markerfacecolor='none',markeredgecolor='red',markeredgewidth=2)
plt.xlim([-6,6])
plt.ylim([-6,6.5])
plt.legend()
plt.title('HMC for sampling a bivariate normal distribution')
```
